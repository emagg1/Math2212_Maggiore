# Aggregating Data
How can we use sum, average, or otherwise aggregate data that shares a similar property?

**Standard aggregating functions from numpy**
count() - this function counts how many values there are

sum() - adds the values of interest 

mean() - finds the mean of the aggregate values

median()

min() - finds the minimum values 

max() - finds the maximum value

prod() - the multiplies all the values

std() - returns the standard  deviation of all the values 

var() - returns the variance of all the values 

```python
# finds the mean Value within each country
group_mean = edu[['GEO', 'Value']].groupby('GEO').mean()
print group_mean
```

```python
group_max = edu[['TIME', 'Value']].groupby('TIME').max()
print group_max
```

# Pivot Tables
A pivot table is similar to these tables that we got using .groupby() except instead of aggregating over one variable, we aggregate over two variables.

```python
ptable = pd.pivot_table(
    edu,
    values = 'Value',
    index = 'TIME',
    columns = 'GEO',
    aggfunc = np.sum
)

print ptable
```

```python
# Worksheet 8 (#6 on Canvas) Print a pivot table that has the countries for 
# rows and year for columus and places the mean value in each of the tables cells
ptable = pd.pivot_table(
    edu,
    values = 'Value',
    index = 'GEO',
    columns = 'TIME',
    aggfunc = np.mean
)

ptable
```
