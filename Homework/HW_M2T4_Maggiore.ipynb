# Question 4 
Two ways outliers may come to exist in datasets is if maybe the people proctering the data are biased with their participants, or maybe they gave this data in an area with the lack of a specific age, or an area that has lower income with an acception of those very rich outliers. A better way to avoid outliers is to make sure you're not being biased with your data and allow multiple forms of that outlier the chance to also participate. Their also may be outliers that you just can't help. There may be data of a millionare in the data, where obviously that would be an obvious outlier that not everyone will be able to match.

```python
# Boxplot prior to dropping outliers
loans.boxplot(column = 'debt_to_income')
```

```python
# create a new dataframe with the outliers dropped 
loans_expert_outlier_drop = loans.drop(
    loans.index[
        (loans['debt_to_income'] >150)    
    ]
)
loans_expert_outlier_drop
```

```python
loans_expert_outlier_drop.boxplot(column = 'debt_to_income')
```

```python
# Dropping outliers using the mean method 
d_to_i = loans['debt_to_income']
d_to_i_mean = loans['debt_to_income'].mean()
d_to_i_std = loans['debt_to_income'].std()

# count the number of outliers we have 
len( loans[ d_to_i > d_to_i_mean + 2.5*d_to_i_std ])
len( loans[ d_to_i < d_to_i_mean - 2.5*d_to_i_std ])

# count the number of outliers we have from high income earners
len( loans[ (d_to_i > 300 ) & ( d_to_i > d_to_i_mean + 2.5*d_to_i_std ) ])
len( loans[ (d_to_i < 200) & ( d_to_i < d_to_i_mean - 2.5*d_to_i_std ) ])

# drop the outliers
loans_expert_outlier_drop_2 = loans.drop(
    loans.index[
        (d_to_i > d_to_i_mean + 2.5*d_to_i_std) | (d_to_i < d_to_i_mean - 2.5*d_to_i_std )
        
    ]
)

loans_expert_outlier_drop_2
```

```python
# Median method
d_to_i = loans['debt_to_income']
d_to_i_median = loans['debt_to_income'].median()
d_to_i_iqr = loans['debt_to_income'].quantile(0.75) - loans['debt_to_income'].quantile(0.25)

# count the number of outliers we have 
len( loans[ d_to_i > d_to_i_median + 1.5*d_to_i_iqr ])
len( loans[ d_to_i < d_to_i_median - 1.5*d_to_i_iqr ])

# count the number of outliers we have from high income earners
len( loans[ (d_to_i < 35) & ( d_to_i > d_to_i_median + 1.5*d_to_i_iqr ) ])
len( loans[ (d_to_i > 250) & ( d_to_i < d_to_i_median - 1.5*d_to_i_iqr ) ])

# drop the outliers
loans_median_outlier_drop_3 = loans.drop(
    loans.index[
        (d_to_i > d_to_i_median + 1.5*d_to_i_iqr) | (d_to_i < d_to_i_median - 1.5*d_to_i_iqr )
        
    ]
)

print d_to_i > d_to_i_median + 1.5*d_to_i_iqr
print d_to_i < d_to_i_median - 1.5*d_to_i_iqr
loans_median_outlier_drop_3
```
