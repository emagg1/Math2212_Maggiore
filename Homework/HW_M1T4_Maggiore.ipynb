# Question 1
Define a piecewise function using an "if, elif, else" statement

```python
# Example of a piecewise function using "if, elif"
Age = 27
if Age >= 60:
    print ('Senior Discount')
elif 18 <= Age < 60:
    print('No Discount')
else:
    print ('Junior Discount')
```

```python
# return the rows with no missing data
census[census['total_family_income'].notnull()]
census[census['total_personal_income'].notnull()]
```

# Question 4

```python
census[census['sex'] == 'Male'] [["sex", "total_family_income"]]
census.sort_values(by = 'total_family_income', ascending = True, inplace = True )
census[census['sex'] == 'Male'] [["sex", "total_family_income"]]
```

```python
census[census['sex'] == 'Female'] [["sex", "total_family_income"]]
census.sort_values(by = 'total_family_income', ascending = False, inplace = True )
census[census['sex'] == 'Female'] [["sex", "total_family_income"]]
```
