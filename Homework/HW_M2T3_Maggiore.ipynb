```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
census = pd.read_csv('census.csv',
                  na_values = ':',
                  usecols = [ 'census_year', 'state_fips_code', 'age',
                              'sex', 'race_general', 'marital_status', 'total_personal_income'])
census
```

##### 1. There are 500 rows or observations included in this census dataframe. 

 #### 2. Variables in the dataframe are 'census_year', 'state_fips_code', 'age', 'sex', 'race_general', 'marital_status', and  'total_personal_income'.

```python
# Outputting only numerical values throughout the data, sorts out which columns are numerical 
census.select_dtypes(include = ['number'])
# Therefore, the numerical variables are census_year, age, and total_personal_income.
```

```python
# Outputting only categorical values throughout the data, sorts out which columns are objects  
census.select_dtypes(include = ['object'])
# Therefore, the categorical objects of the dataframe are state_fips_code, sex, race_general_ and marital_status. 
```
