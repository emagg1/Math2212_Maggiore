```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
census = pd.read_csv('census.csv',
                  na_values = ':',
                  usecols = [ 'census_year', 'state_fips_code', 'age',
                              'sex', 'race_general', 'marital_status', 'total_personal_income'])
census
```

```python
# Finding the dimensions of the dataset using shape 
census.shape
```

```python
# Finding characteristics of data using info()
census.info()
```

```python
# Showing basic statistical characteristics of each numerical feature 
census.describe()
```

```python
# Handeling missing values in the dataset by dropping missing values
census.dropna
census
```

```python
# Checking for the sum of missing values of each column 
census.isnull().sum()
```

```python
# Replacing missing data with a 0 
census.fillna(0)
```

```python
# Checking for duplicate values in the dataset using drop.duplicates()
census.drop_duplicates()
census
```

```python
# Handling outliers in data 
census.boxplot(column = 'total_personal_income')   
```

```python
# Seeing outlier in two groups, female/male ages for example 
census.boxplot(column = 'age', by = 'sex')  
```

```python
census.boxplot(column = 'total_personal_income', by = 'age')  
```

```python
census.boxplot(column = 'total_personal_income', by = 'sex')  
```

```python
census.boxplot(column = 'age')  
```

```python
# Finding the correlation among the columns using Pearson method 
census.corr(method = 'pearson')
```

```python
# Finding correlations among the columns using the Kendall method
census.corr(method = 'kendall')
```

```python
# Finding correlations among the columns using the Spearman method
census.corr(method = 'spearman')
```

```python
# Sampling the data, which takes a sample of 6
census.sample(6)
```

```python
# Sample of 1
census.sample(1)
```

```python
# If you want to see only numerical outputs in the data
census.select_dtypes(include = ['number'])
```

```python
# If you want to see only objects 
census.select_dtypes(include = ['object'])
```

```python
census.total_personal_income.max()
```

```python
# Which person and their information had this value for their total personal income?
census[census.total_personal_income == census.total_personal_income.max()]
```

```python
# Same experimentation with age, highest age but with all of their information and row number 
census[census.age == census.age.max()]
```

```python
# Creating two histograms of the same goal only separating them by sex 
census['total_personal_income'].hist(by = census['sex'])
```

```python
# Experimenting with Seaborn
import seaborn as sns
sns.boxplot(x = 'sex', y = 'total_personal_income', data = census)
```

```python
# Further Seaborn experimentation 
sns.boxplot(x = 'sex', y = 'total_personal_income', data = census)
sns.stripplot(x = 'sex', y = 'total_personal_income', data = census,
              jitter = True,
              marker = 'o',
              alpha = 0.8,
              color = "pink")
```

```python
# Experimenting with heatmaps 
corr = census.corr()
corr
```

```python
sns.heatmap(corr)
```

```python
# Adjusting figure size of the heatmap 
fig, ax = plt.subplots(figsize = (20,30))
sns.heatmap(corr, square = False)
```
