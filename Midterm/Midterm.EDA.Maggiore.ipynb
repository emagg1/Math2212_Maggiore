```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
census = pd.read_csv('census.csv',
                  na_values = ':',
                  usecols = [ 'census_year', 'state_fips_code', 'age',
                              'sex', 'race_general', 'marital_status', 'total_personal_income'])
census
```

```python
# Finding the dimensions of the dataset using shape 
census.shape
```

```python
# Finding characteristics of data using info()
census.info()
```

```python
# Showing basic statistical characteristics of each numerical feature 
census.describe()
```

```python
# Handeling missing values in the dataset by dropping missing values
census.dropna
census
```

```python
# Checking for the sum of missing values of each column 
census.isnull().sum()
```

```python
# Replacing missing data with a 0 
census.fillna(0)
```

```python
# Checking for duplicate values in the dataset using drop.duplicates()
census.drop_duplicates()
census
```

```python
# Handling outliers in data 
census.boxplot(column = 'total_personal_income')   
```

```python
# Experimenting with Seaborn
import seaborn as sns
sns.boxplot(x = 'sex', y = 'total_personal_income', data = census)
```

```python
# Further Seaborn experimentation, making the individual data show but in pink because its my favorite color
sns.boxplot(x = 'sex', y = 'total_personal_income', data = census)
sns.stripplot(x = 'sex', y = 'total_personal_income', data = census,
              jitter = True,
              marker = 'o',
              alpha = 0.8,
              color = "pink")
```

```python
# Individual data shown in black in case pink was hard to see 
sns.boxplot(x = 'sex', y = 'total_personal_income', data = census)
sns.stripplot(x = 'sex', y = 'total_personal_income', data = census,
              jitter = True,
              marker = 'o',
              alpha = 0.8,
              color = "black")
```
