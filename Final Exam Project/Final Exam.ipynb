```python
# Part A, Task 4
def  median_bootstrap(california, texas, number_of_resamples):
    x = [0]* number_of_resamples #initialize  an  array 
    for i in  range(number_of_resamples):
        resample = [california[j]
                    for j
                    in np.random.randint(len(california), size=len(texas))
                   ]
        x[california.median()- texas.median()] = np.median(resample)
    return x
```

```python
m = median_bootstrap(california, texas, 100)
plt.hist(m)
plt.show()
```

```python
# Part B, Task 4
def  mean_bootstrap2(census['total_family_income'], number_of_resamples):
    x = [0]* number_of_resamples #initialize  an  array 
    for p in  range(number_of_resamples):
        resample = [census['total_family_income'][j]
                    for j
                    in np.random.randint(len(census['total_family_income']), size=len(census['total_family_income']))
                   ]
        x[p15 - p100] = np.mean(resample)
    return x
```

```python
m = mean_bootstrap2(census['total_family_income'], 100)
plt.hist(m)
plt.show()
```

```python
# Part C, Task 4
def  mean_bootstrap3(age_ny, age_florida, number_of_resamples):
    x = [0]* number_of_resamples #initialize  an  array 
    for i in  range(number_of_resamples):
        resample = [age_ny[j]
                    for j
                    in np.random.randint(len(age_ny), size=len(age_florida))
                   ]
        x[age_florida.mean()- age_ny.mean()] = np.mean(resample)
    return x
```

```python
m = mean_bootstrap3(age_ny, age_florida, 100)
plt.hist(m)
plt.show()
```

```python
# Part D, Task 4
def  std_bootstrap(florida, newyork, number_of_resamples):
    x = [0]* number_of_resamples #initialize  an  array 
    for i in  range(number_of_resamples):
        resample = [florida[j]
                    for j
                    in np.random.randint(len(florida), size=len(newyork))
                   ]
        x[florida.std()- newyork.std()] = np.std(resample)
    return x
```

```python
m = std_bootstrap(florida, newyork, 100)
plt.hist(m)
plt.show()
```
